```python
import string
# string.ascii_lowercase
cipher ={}
for idx, letter in enumerate(string.ascii_lowercase):
    cipher[letter] = string.ascii_lowercase[(idx+2)%26]  # %len helps us avoid index errors, but is a costly calc
cipher
```

```python
sentence = "g fmnc wms bgblr rpylqjyrc gr zw fylb. rfyrq ufyr amknsrcpq ypc dmp. bmgle gr gl zw fylb gq glcddgagclr ylb rfyr'q ufw rfgq rcvr gq qm jmle. sqgle qrpgle.kyicrpylq() gq pcamkkclbcb. lmu ynnjw ml rfc spj."
```

```python
# one approach, construct a new string based on the cipher, is likely performant
new_sentence = ""
for character in sentence:
    next_character = cipher.get(character, character)  # get the value, unless it isn't there, use character
    new_sentence += next_character
new_sentence
```

```python
# make trans! string.maketrans()
```

```python
made_trans = sentence.maketrans(cipher)
sentence.translate(made_trans)
```

```python
# now to do it on the url
url_substring = "map"
url_substring.translate(made_trans)
```

```python
response = requests.get("http://www.pythonchallenge.com/pc/def/equality.html")
```

```python
soup = BS(response.text, 'html.parser')
```

```python
comments = soup.find_all(string=lambda text: isinstance(text, Comment))  # https://stackoverflow.com/questions/33138937/how-to-find-all-comments-with-beautiful-soup
```

```python
comment = comments[0]
```

```python
shifting_window = comment[0:9]  # make the first shifting window aAAAaAAAa should be what we look for. 
```

```python
def is_candidate(shifting_window:list) ->bool:
    return shifting_window[0].islower() and shifting_window[1].isupper() and shifting_window[2].isupper() \
            and shifting_window[3].isupper() and shifting_window[4].islower() and shifting_window[5].isupper() \
            and shifting_window[6].isupper() and shifting_window[7].isupper() and shifting_window[8].islower()
```

```python
for idx in range(len(comment)):
    shifting_window = comment[idx:idx+9]  # no need to bother with index out of range error handling. That means it's at an end anyway
    if is_candidate(shifting_window):
        print(shifting_window[4]) # want to print the small character within!
```

```python
def parse_response(response_text:str, last_slug:int)->int:
    """
    Processes a string and returns the integer to add to the next url args.
    https://stackoverflow.com/questions/4289331/how-to-extract-numbers-from-a-string-in-python
    """
    print(response_text)
    if response_text == "Yes. Divide by two and keep going.":
        return last_slug/2
    response_text = response_text.split(" ")  # since there may be mulitple numbers, and we only want the last
    return int(''.join(filter(str.isdigit, response_text[-1])))
```

```python
# f"http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing={slug}"
```

```python
slug = 82682 
last_slug = 0
break_counter = 0
break_max = 400
while True:
    if break_counter >= break_max:
        break
    response = requests.get(f"http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing={slug}")
    last_slug = slug
    slug = parse_response(response.text, last_slug)
    break_counter +=1
    
```

```python
# Now with the image downloaded, what next?
# get the grey bar values?
# Pillow!
```

```python
from PIL import Image
```

```python
image = Image.open('oxygen.png')
```

```python
image.width
```

```python
image.height
```

```python
# grey should be in the middle, right?
# image[0:image.height/2]

image.getpixel((0,(image.height/2)))
```

```python
row = [image.getpixel((idx, image.height/2)) for idx in range(image.width)]
```

```python
row
```

```python
# so all the duplicated ones are what we want, I bet
# looks like after the first one they repeat every 7

de_duplicated_row = row[::7]
```

```python
de_duplicated_row
```

```python
# only want the ones where all RGB match
# and may as well only get the single value while we're at it
prepared_row = []
for pixel in de_duplicated_row:
    if pixel[0] == pixel[1] == pixel[2]:
        prepared_row.append(pixel[0])
```

```python
prepared_row
```

```python
# now to convert it into a string
final_sentence = ""
for char in prepared_row:
    final_sentence +=  str(chr(char))
```
